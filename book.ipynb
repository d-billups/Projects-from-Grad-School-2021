(5 points) You will be using part of the Goodbooks 10k dataset for this module. To start, complete the function and read the data/goodreads-books.csv file into a list. Your output should be a list of dictionaries (one for for each book), which each contain the following fields: 'authors', original 'title', original 'publication year', 'average rating', and 'ratings count'. Note: you must convert the average rating and ratings count into float and int types, respectively.

```python
import csv
import pandas as pd
import regex as re
```

```python
books = csv.reader(open('/Users/diorbillups/DSCI_511_Assignments/Good_books_assignment/data/goodreads-books.csv','r'))
book = list(books)
```

```python
header = book[0]
```

```python
header
```

```python
#Grabbing the indices for all of the categories we 
```

```python
auth = header.index('authors')
orig_title = header.index('original_title')
orig_pubyear = header.index('original_publication_year')
avg_r = header.index('average_rating')
rating_count = header.index('ratings_count')
```

```python
l = []

for i in book[1:]:
    info = {}
    rating_c = int(float(i[rating_count]))
    avg = float(i[avg_r])
    py = i[orig_pubyear]
    j = ''.join(re.findall('\d{4}',py))
    
    info['author']=i[auth]
    info['original_title']=i[orig_title]
    info['original_pub_year']=j
    info['avg_rating']=avg
    info['ratings_count']=rating_c
    
    l.append(info)
```

(5 points) Now complete the below function to take two arguments: the list, and an integer value for minimum ratings count. The new function should now sort and filter the list, returning a list of books sorted by average rating that have been rated by more than a specified number of users. Here, you must use an approach using loops and if statements—in subsequent parts, comprehensions and the built-in filter() must be utilized instead (look up documentation and examples).

Syntax: filter(function, sequence)

function: function that tests if each element of a sequence is true or not.
sequence: sequence which needs to be filtered, it can be sets, lists, tuples, or containers of any iterators.

```python
def sort_ratings(listy,num):
    y = sorted(listy, key = lambda x: x['ratings_count'],reverse=True)
    z = []
    for i in y:
        if i['ratings_count'] > num:
            z.append(i)
            
    return z
```

```python
j = sort_ratings(l,1000000)[:5]
```

First, use a list comprehension inside of the sorted() function:

```python
def sort_filter(listx,num):
    x = sorted([i for i in listx if i['ratings_count']>num], key = lambda x: x['avg_rating'],reverse=True)
    return x
```

```python
sort_filter(l,1000000)[:5]
```

```python
filter_sort = list(filter(lambda x: x['ratings_count'] > 1000000, l))
```
